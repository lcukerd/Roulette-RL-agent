```python
import numpy as np
import pandas as pd
import time
import gym
import random
from IPython.display import clear_output
```

```python
state_size = 0
action_size = 0
ALPHA = 0.1
Gamma = 0.9
show = False
np.random.seed(2)
```

```python
def choose_action(q_table,S,Greed_F):
    q = q_table[S,:]
    if ((np.random.uniform() < Greed_F) or ((q == 0).all())) and (Frame_Time != 1):
        a = random.randint(0,3)
    else:
        a = q.argmax()
    return a
```

```python
def create_q_table(states):
    q_table = np.zeros((states,action_size))
    return q_table
```

```python
def simulate(q_table,Greed_F):
    env = gym.make('FrozenLake8x8-v0')
    for gen in range(Episodes):
        steps = S = R = 0
        env.reset()
        Terminated = False
        while not Terminated:
            A = choose_action(q_table,S,Greed_F)
            if (show is True):
                env.render()
                time.sleep(Frame_Time)
                clear_output()
            if (Greed_F>0.5):
                Greed_F *= 0.99999
            S_,R,Terminated,_ = env.step(A)
            LEFT = 0
            DOWN = 1
            RIGHT = 2
            UP = 3
#             if S_-S == 8:
#                 A = 1
#             elif S_-S == 1:
#                 A = 2
#             elif S_-S == -1:
#                 A = 0
#             else:
#                 A =3
            if Terminated is True and R == 0:
                R = -1
            predict_q = q_table[S,A]
            if (Terminated != True):
                target_q = R + Gamma*q_table[S_,:].max()
            else:
                target_q = R
            q_table[S,A] += ALPHA * (target_q - predict_q)
            S = S_
            steps += 1
        if R == 1:
            print("Episode: {}/{}, e: {:.2}, Steps: {}, Reward: {}"
              .format(gen, Episodes, Greed_F, steps,R))
    env.close()
    return q_table
```

```python
np.random.seed(2)
Frame_Time = 2
Episodes = 10000
state_size = 8*8
action_size = 4
show = False
q_table = create_q_table(state_size)
q_table = simulate(q_table,1.0)
print (q_table)
```
